```python
# 1. What is the difference between a function and a method in Python?
# A function is a block of reusable code defined outside any class. It can be called directly.
# A method is a function associated with an object and is called on the object.
# Example:
def function_example():
    return "This is a function."

class ExampleClass:
    def method_example(self):
        return "This is a method."

# Function call
print(function_example())

# Method call
obj = ExampleClass()
print(obj.method_example())

# 2. Explain the concept of function arguments and parameters in Python.
# Parameters are variables listed in a function's definition.
# Arguments are the actual values passed to a function when it is called.
# Example:
def greet(name):  # 'name' is a parameter
    return f"Hello, {name}!"

# Calling the function with an argument
print(greet("Alice"))  # "Alice" is the argument

# 3. What are the different ways to define and call a function in Python?
# Functions can be defined using the 'def' keyword or as lambda functions. They can be called by their name followed by parentheses.
# Example:
# Standard function
def add(a, b):
    return a + b

# Lambda function
add_lambda = lambda a, b: a + b

# Calling both
print(add(2, 3))       # Output: 5
print(add_lambda(2, 3))  # Output: 5

# 4. What is the purpose of the `return` statement in a Python function?
# The 'return' statement is used to send a result back to the caller from a function.
# Example:
def square(number):
    return number ** 2

result = square(4)
print(result)  # Output: 16

# 5. What are iterators in Python and how do they differ from iterables?
# Iterables are objects that can be looped over, like lists or strings. Iterators are objects obtained from iterables using the 'iter()' function and produce values one at a time using 'next()'.
# Example:
my_list = [1, 2, 3]  # Iterable
iterator = iter(my_list)  # Iterator

print(next(iterator))  # Output: 1
print(next(iterator))  # Output: 2

# 6. Explain the concept of generators in Python and how they are defined.
# Generators are functions that yield values one at a time using the 'yield' keyword instead of 'return'.
# Example:
def count_up_to(max):
    count = 1
    while count <= max:
        yield count
        count += 1

for num in count_up_to(3):
    print(num)  # Output: 1, 2, 3

# 7. What are the advantages of using generators over regular functions?
# Generators use less memory as they yield values lazily one at a time instead of storing them in memory.
# Example:
def generator_example():
    for i in range(5):
        yield i  # Lazy evaluation

for value in generator_example():
    print(value)  # Outputs: 0, 1, 2, 3, 4

# 8. What is a lambda function in Python and when is it typically used?
# Lambda functions are anonymous functions defined using the 'lambda' keyword. They are used for small, short-term tasks.
# Example:
square = lambda x: x ** 2
print(square(5))  # Output: 25

# 9. Explain the purpose and usage of the `map()` function in Python.
# The 'map()' function applies a given function to each item in an iterable and returns a map object.
# Example:
numbers = [1, 2, 3]
squared = map(lambda x: x ** 2, numbers)
print(list(squared))  # Output: [1, 4, 9]

# 10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
# - 'map()' applies a function to all elements in an iterable.
# - 'reduce()' reduces an iterable to a single value using a function (from functools module).
# - 'filter()' filters elements in an iterable based on a function returning True or False.
# Example:
from functools import reduce

nums = [1, 2, 3, 4]
print(list(map(lambda x: x ** 2, nums)))  # map: [1, 4, 9, 16]
print(reduce(lambda x, y: x + y, nums))  # reduce: 10
print(list(filter(lambda x: x % 2 == 0, nums)))  # filter: [2, 4]

#  11 Using pen & Paper write the internal mechanism for sum operation using  reduce function on this given
#  list:[47,11,42,13];

#SIR I DO NOT KNOW HOW TO INSERT AN IMAGE IN GOOGLE COLAB SO HERE IS MY ANSWER

# Importing reduce from functools
from functools import reduce

# List to perform sum operation on
numbers = [47, 11, 42, 13]

# Applying the reduce function to calculate the sum
sum_result = reduce(lambda x, y: x + y, numbers)

# Displaying the result
print(sum_result)


#1. **List**: [47, 11, 42, 13]

#2. **Step-by-Step Mechanism**:

 #   - Take the first two elements: 47 and 11
  #   - Perform the operation: `47 + 11 = 58`
   #  - Result after the first iteration: `58`

   #- **Second Iteration**:
    # - Take the result from the first iteration (`58`) and the next element (`42`)
     #- Perform the operation: `58 + 42 = 100`
     #- Result after the second iteration: `100`

   #- **Third Iteration**:
    # - Take the result from the second iteration (`100`) and the next element (`13`)
     #- Perform the operation: `100 + 13 = 113`
     #- Result after the third iteration: `113`

#3. **Final Output**: `113`

#PRACTICAL QUESTIONS

# 1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

def sum_of_even_numbers(numbers):
    return sum(num for num in numbers if num % 2 == 0)

# Example:
numbers = [1, 2, 3, 4, 5, 6]
sum_of_even_numbers(numbers)  # Output: 12


# 2. Create a Python function that accepts a string and returns the reverse of that string.

def reverse_string(s):
    return s[::-1]

# Example:
reverse_string("hello")  # Output: "olleh"


# 3. Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.

def square_numbers(numbers):
    return [num ** 2 for num in numbers]

# Example:
square_numbers([1, 2, 3, 4])  # Output: [1, 4, 9, 16]


# 4. Write a Python function that checks if a given number is prime or not from 1 to 200.

def is_prime(n):
    if n < 2:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

# Example:
is_prime(7)  # Output: True
is_prime(10)  # Output: False


# 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

class FibonacciIterator:
    def __init__(self, n):
        self.n = n
        self.a, self.b = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count < self.n:
            value = self.a
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return value
        else:
            raise StopIteration

# Example:
fib_iter = FibonacciIterator(5)
for num in fib_iter:
    print(num)  # Output: 0, 1, 1, 2, 3


# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

def powers_of_2(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

# Example:
list(powers_of_2(4))  # Output: [1, 2, 4, 8, 16]


# 7. Implement a generator function that reads a file line by line and yields each line as a string.

def read_file_line_by_line(filename):
    with open(filename, 'r') as file:
        for line in file:
            yield line.strip()

# Example:
# Call the function with a file path (replace 'filename.txt' with an actual file)
# for line in read_file_line_by_line('filename.txt'):
#     print(line)


# 8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

tuples = [(1, 'one'), (3, 'three'), (2, 'two')]
sorted_tuples = sorted(tuples, key=lambda x: x[1])

# Example:
sorted_tuples  # Output: [(3, 'three'), (1, 'one'), (2, 'two')]


# 9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

celsius_temps = [0, 20, 37, 100]
fahrenheit_temps = list(map(celsius_to_fahrenheit, celsius_temps))

# Example:
fahrenheit_temps  # Output: [32.0, 68.0, 98.6, 212.0]


# 10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

def remove_vowels(s):
    return ''.join(filter(lambda x: x not in 'aeiouAEIOU', s))

# Example:
remove_vowels("Hello World")  # Output: "Hll Wrld"

#ANSWER 11

#  Python program using lambda and map to solve the problem based on the provided accounting routine:

# Input list of orders
orders = [
    [34587, 'Learning Python, Mark Lutz', 40.95, 4],
    [98762, 'Programming Python, Mark Lutz', 56.80, 5],
    [77226, 'Head First Python, Paul Barry', 32.95, 3],
    [88112, 'Einführung in Python3, Bernd Klein', 24.99, 3]
]

# Using lambda and map to calculate the total price for each order
def calculate_order_total(order):
    order_number, book_title, price_per_item, quantity = order
    total = price_per_item * quantity
    if total < 100:
        total += 10  # Add 10 if the order total is less than 100
    return (order_number, total)

# Apply the function to the orders list using map
order_totals = list(map(lambda order: calculate_order_total(order), orders))

# Output the result
order_totals

```
